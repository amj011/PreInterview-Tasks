```python
import pandas_datareader as data 
from datetime import date
start = '05-05-2021'
end = date.today()
df = data.DataReader('RELIANCE.NS', 'yahoo',start, end )
df.head()
```

```python
df
```

```python
df.to_csv('final_df')
```

```python
df = df.drop(['High', 'Low','Open','Volume','Adj Close'], axis = 1)
df.head()
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
df.Close.plot(figsize=(7,5),use_index=False)
#plt.plot(df.Date,df.Close)
plt.grid()
plt.ylabel("Price in Rupees")
plt.show()
```

```python
SMA20 = df.Close.rolling(20).mean()
SMA20
# For the First 20 days there will be no Data
```

```python

plt.figure(figsize = (12,6))
plt.ylabel("Price in Rupees")
plt.xlabel("Dates")
plt.plot(df.Close)
plt.plot(SMA20, 'r')
```

```python
SMA50 = df.Close.rolling(50).mean()
SMA50
# For the First 50 days there will be no Data
```

```python
plt.figure(figsize = (12,6))
plt.plot(df.Close)
plt.ylabel("Price in Rupees")
plt.xlabel("Dates")
plt.plot(SMA50, 'r')
```
